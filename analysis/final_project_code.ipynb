## 1. Imports

```python
import os
import os.path as pt
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re
import scipy

from sklearn.model_selection import RepeatedKFold

import statsmodels.api as sm
import statsmodels.formula.api as smf
```

## 2. Data Loading

```python
ROOT_DIR = '../'
EXP_FORMAT = '\d+g_*' # matches [number]g_[exp]

fs = 100 # sampling freq
```

```python
# Load data

experiment_list = [x for x in os.listdir(ROOT_DIR) if re.match(EXP_FORMAT, x)]
experiment_meta = {}

for exp_dir in experiment_list:  
    weight, name = exp_dir.split('g_')
    weight = float(weight)
    
    exp_path = pt.join(ROOT_DIR, exp_dir)
    
    csv_paths = [os.path.join(exp_path, x) for x in os.listdir(exp_path) if x.endswith('csv')]
    csv_paths.sort()
    
    experiment_meta[name] = {
        'weight': weight,
        'csv_paths': csv_paths
    }
```

```python
experiment_meta
```

```python
def load_obj(obj, ind, subtract_mean=True, filter_freq=None):
    data = pd.read_csv(experiment_meta[obj]['csv_paths'][ind], header=None)
    
    # Translate the signal such that it has zero mean
    if subtract_mean:
        data = data - np.asarray(data.mean(axis=1)).reshape(-1, 1)
    
    # Apply Gaussian filter
    if filter_freq:
        data = pd.DataFrame(scipy.ndimage.gaussian_filter1d(np.asarray(data), filter_freq))
    
    return data
```

## 3. Data Analysis

The general hypothesis behind our project is that heavier objects should reduce the phone's vibrations; thus, we plot the absolute amplitudes for the vibrations of a few objects.

The absolute amplitude/acceleration is $|x_i - \bar{x}|$, where $x_i$ are the signal measurements. This measures the displacement from the mean value of the signal.

The MAA: mean absolute acceleration is $\overline{x_i - \bar{x}} = \frac{\sum_{i=1}^{n}|x_i - \bar{x}|}{n}$, where $n$ is the number of measurements i.e. 1000.

```python
def plot_abs_wave(obj2, window=None, obj1='baseline', i1=0, i2=0, filter_freq=None):
    obj1_data = load_obj(obj1, i1, filter_freq=filter_freq)
    obj2_data = load_obj(obj2, i2, filter_freq=filter_freq)    
    
    if window is None:
        le, re = 0, obj1_data.shape[1]
    else:
        le, re = window
    
    fig, axes = plt.subplots(3, 1, sharex=True, figsize=(15, 8))
    fig.suptitle(f'{obj1} vs {obj2} (weight {experiment_meta[obj2]["weight"]}g)')
    
    for i, v in enumerate(['x', 'y', 'z']):
        axes[i].plot(np.abs(obj1_data.loc[i, le:re]), color='blue', linestyle='dotted', alpha=0.5, label=obj1)
        axes[i].plot(np.abs(obj2_data.loc[i, le:re]), color='orange', linestyle='dotted', alpha=0.5, label=obj2)     
        
        axes[i].plot(list(range(le, re)), np.repeat(np.abs(obj1_data.loc[i]).mean(), re-le), color='blue')
        axes[i].plot(list(range(le, re)), np.repeat(np.abs(obj2_data.loc[i]).mean(), re-le), color='orange')
        
        axes[i].set_xlabel(v)
        
    fig.tight_layout()
    plt.legend()
```

In the above plots, the mean absolute acceleration (MAA) is shown as the solid line and the absolute accelerations are shown as a dotted line.

We observe:
- Across all but the z-axis vibrations of the camera (final axis, final plot), the baseline vibrations are more than the object vibrations, which is as we expect according to the hypothesis.
- The data is quite noisy - this prompts us to experiment with filtering later in the notebook.
- z-axis seems to have minimum difference between the baseline and the object mean absolute amplitudes, which is quite unintuitive, as that is the direction in which the object is kept. Possible reasons - something to do with gravity? maybe phone case?

In any case, this suggests creating a feature given by the mean absolute acceleration of vibration in each direction i.e x_MAA, y_MAA, z_MAA.

```python
# Generate a dataframe containing the mean absolute amplitude for each object's measurements
def generate_feature_df(filter_freq=None):
    records = []

    for exp_name, exp in experiment_meta.items():
        for i in range(3):
            data = load_obj(exp_name, i, filter_freq=filter_freq)
            x, y, z = data.abs().mean(axis=1)

            records.append({
                'name': exp_name,
                'weight': exp['weight'],
                'x': x,
                'y': y,
                'z': z
            })
    return pd.DataFrame(records).sort_values(by='weight').reset_index(drop=True)

feature_df = generate_feature_df()
```

## 4. Modelling

### 4.1 Simple linear regression

Let's now fit a simple linear regression model: $\text{weight} = \beta_0 + \beta_{1}\text{x-MAA}+ \beta_{2}\text{z-MAA} + \epsilon$. y is excluded for collinearity reasons.

```python
model_1 = smf.ols('weight ~ x + z', data=feature_df).fit()
model_1.summary()
```

```python
model_1.params
```

As expected, we get a negative coefficient for x and a positive coeff. for z. Both coefficients are significant, and the R-squared is quite high. Let's see the predicted values:

```python
def evaluation_df(feature_df, predictors, model, results_only=True):
    results = pd.DataFrame()
    results['name'] = feature_df['name']
    results['weight_true'] = feature_df['weight']
    results['weight_pred'] = model.predict(feature_df[predictors])
    results['relative_error'] = np.abs(results['weight_true'] - results['weight_pred']) / results['weight_true']

    if results_only:
        return results.relative_error.replace([np.inf, -np.inf], np.nan).mean()
    
    return results

evaluation_df(feature_df, ['x', 'z'], model_1, results_only=False)
```

```python
print('Mean Relative Error:', evaluation_df(feature_df, ['x', 'z'], model_1))
print('Mean Relative Error after excluding low weight objects:', evaluation_df(feature_df, ['x', 'z'], model_1, results_only=False).iloc[12:, -1].mean())
```

We immediately see poor performance (negative preds) for the extremely low weight objects (<10g), suggesting a potential limitation - maybe the difference in vibration magnitude is not high enough for such light objects.

The mean relative error is quite high at 85%; however, if the lightweight objects are excluded (coin, credit card, grape), then the mean relative error drops to 39%. This again suggests that we may wish to either **exclude lightweight objects**, or consider constructing a separate model for them.

### 4.2 Linear regression with interactions

Another model to try is the model with interactions i.e. $\text{weight} = \beta_0 + \beta_{1}\text{x-MAA}+ \beta_{2}\text{z-MAA} + \beta_{3}\text{x-MAA*z-MAA}+ \epsilon$.

```python
model_2 = smf.ols('weight ~ x*z', data=feature_df).fit()
model_2.summary()
```

We see a moderate improvement in R-squared; coefficients are still significant.

```python
print('Mean Relative Error:', evaluation_df(feature_df, ['x', 'z'], model_2))
print('Mean Relative Error after excluding low weight objects:', evaluation_df(feature_df, ['x', 'z'], model_2, results_only=False).iloc[12:, -1].mean())
```

There is a considerable drop in the MRE.

### 4.3 Linear regression with log-transform on weight

Another model to try is the model with log-transform on the weight i.e. $\text{log(weight)} = \beta_0 + \beta_{1}\text{x-MAA}+ \beta_{2}\text{z-MAA} + \epsilon$.

```python
feature_df['log_weight'] = np.log(feature_df['weight'])
feature_df.head()
```

```python
model_3 = smf.ols('log_weight ~ x+z', data=feature_df.loc[3:,:]).fit()
model_3.summary()
```

```python
model_3.params
```

```python
def evaluation_df_log(feature_df, predictors, model, results_only=True):
    results = pd.DataFrame()
    results['name'] = feature_df['name']
    results['weight_true'] = feature_df['weight']
    results['weight_pred'] = np.exp(model.predict(feature_df[predictors]))
    results['relative_error'] = np.abs(results['weight_true'] - results['weight_pred']) / results['weight_true']
    
    if results_only:
        return results.relative_error.replace([np.inf, -np.inf], np.nan).mean()

    return results
```

```python
print('Mean Relative Error:', evaluation_df_log(feature_df, ['x', 'z'], model_3))
print('Mean Relative Error after excluding low weight objects:', evaluation_df_log(feature_df, ['x', 'z'], model_3, results_only=False).iloc[12:, -1].mean())
```

We clearly see that there is a significant drop in Mean Relative Error indicating that low weight objects are being weighed more accurately as compared to the other significant models.
However, the mean relative error for higher weight objects is between that of the previous 2 models.

### 4.4 Model comparison - cross validation on train set

We now compare the performance of the three models.

```python
def cross_validate_models(data, models):
    data = data.reset_index()
    
    cv_scores = {model: [] for model in models.keys()}
    
    for train_idx, test_idx in RepeatedKFold(random_state=42, n_splits=10, n_repeats=5).split(data):
        X_train = data.loc[train_idx]
        X_test = data.loc[test_idx]
        
        for model_name, model_str in models.items():
            model_fit = smf.ols(model_str, data=X_train).fit()
            test_score = evaluation_df(X_test, ['x', 'z'], model_fit) if 'log' not in model_name else evaluation_df_log(X_test, ['x', 'z'], model_fit)
            cv_scores[model_name].append(test_score)
    
    print('CV Scores (MREs, or mean relative errors - lower is better):')
    for model, scores in cv_scores.items():
        print(f'{model}: {np.mean(scores)}')
```

```python
filtered_feature_df = feature_df.loc[(50 <= feature_df.weight) & (200 >= feature_df.weight)]
models = {
        'SLR': 'weight ~ x + z',
        'SLR+interaction': 'weight ~ x*z',
        'SLR+log-transform': 'log_weight ~ x+z'
    }
cross_validate_models(filtered_feature_df, models)
```

We observe that the log-transform model (last model) performs the best, albeit on the train set, evaluated using 10-fold + 5-times-repeated cross validation.

### 4.5 Effect of filtering

We now test the effect of applying a Gaussian filter on the accelerometer readings, using 10-fold + 5-times-repeated cross validation.

```python
def test_gaussian_filter(filter_freq):
    full_data = generate_feature_df(filter_freq=filter_freq)
    filtered_data = full_data.loc[(full_data.weight >= 50) & (full_data.weight <= 200)].reset_index(drop=True)
    filtered_data['log_weight'] = np.log(filtered_data.weight)
    
    cv_scores = []
    
    for train_idx, test_idx in RepeatedKFold(random_state=42, n_splits=10, n_repeats=5).split(filtered_data):
        X_train = filtered_data.loc[train_idx]
        X_test = filtered_data.loc[test_idx]
        
        model = smf.ols('log_weight ~ x+z', data=X_train).fit()
        
        cv_scores.append(evaluation_df_log(X_test, ['x', 'z'], model))
    
    return np.mean(cv_scores), np.std(cv_scores)
```

```python
gaussian_filter_stds = [None] + list(np.logspace(-4, 2, num=7))

best_perf, best_perf_dev, best_perf_std = np.inf, None, None

for std in gaussian_filter_stds:
    perf_mean, perf_std = test_gaussian_filter(std)
    
    print(f"Performance for Gaussian Filter with std. deviation {std}")
    print(f"Mean: {perf_mean:.3f}, std. dev: {perf_std:.3f}")
    print()
    
    if perf_mean < best_perf:
        best_perf = perf_mean
        best_perf_dev = perf_std
        best_perf_std = std
    
print('Best performance:')
print(f"Performance for Gaussian Filter with std. deviation {best_perf_std}")
print(f"Mean: {best_perf:.3f}, std. dev: {best_perf_dev:.3f}")
print()
```

The performance with filtering, using a standard deviation of 10 for the filter is far better than not using the filter. We now fine-grain our filters.

```python
gaussian_filter_stds = np.linspace(0.1, 10, num=20)

best_perf, best_perf_dev, best_perf_std = np.inf, None, None

for std in gaussian_filter_stds:
    perf_mean, perf_std = test_gaussian_filter(std)
    
    print(f"Performance for Gaussian Filter with std. deviation {std}")
    print(f"Mean: {perf_mean:.3f}, std. dev: {perf_std:.3f}")
    print()
    
    if perf_mean < best_perf:
        best_perf = perf_mean
        best_perf_dev = perf_std
        best_perf_std = std
    
print('Best performance:')
print(f"Performance for Gaussian Filter with std. deviation {best_perf_std}")
print(f"Mean: {best_perf:.3f}, std. dev: {best_perf_dev:.3f}")
print()
```

We settle for a standard deviation of about 4.8.

```python
models = {
        'SLR+log-transform+Gaussian-filter': 'log_weight ~ x+z'
    }

full_data = generate_feature_df(filter_freq=4.8)
filtered_data = full_data.loc[(full_data.weight >= 50) & (full_data.weight <= 200)].reset_index(drop=True)
filtered_data['log_weight'] = np.log(filtered_data.weight)

cross_validate_models(filtered_data, models)
```

Thus, we observe that after cross validation, the model using a Gaussian filter + log transform performs the best.

```python
full_data = generate_feature_df(filter_freq=4.8)
filtered_data = full_data.loc[(full_data.weight >= 50) & (full_data.weight <= 200)].reset_index(drop=True)
filtered_data['log_weight'] = np.log(filtered_data.weight)

log_model = smf.ols('log_weight ~ x+z', data=filtered_data).fit()
print(f"Final model performance on train set: {evaluation_df_log(filtered_data, ['x', 'z'], log_model)}")
```

## 5. Evaluation on test set

Now, we can finally test our models on the test sets.

```python
def gen_test_set(filter_freq=False):
    test_set_dir = '../test2'
    test_records = []

    for test_case in os.listdir(test_set_dir):
        test_case_weight, test_case_name = test_case.split('.csv')[0].split('g_')
        test_case_weight = float(test_case_weight)

        data = pd.read_csv(os.path.join(test_set_dir, test_case), header=None)

        # Subtract mean
        data = data - np.asarray(data.mean(axis=1)).reshape(-1, 1)

        # Apply filter
        if filter_freq:
            data = pd.DataFrame(scipy.ndimage.gaussian_filter1d(np.asarray(data), 4.8))
        x, y, z = data.abs().mean(axis=1)

        test_records.append({
            'x': x,
            'y': y,
            'z': z,
            'name': test_case_name,
            'weight': test_case_weight
        })

    test_set = pd.DataFrame(test_records)
    test_set = test_set.loc[(test_set.weight >= 50) & (test_set.weight <=200)]
    
    return test_set

filtered_test_set = gen_test_set(filter_freq=True)
unfiltered_test_set = gen_test_set(filter_freq=False)
```

```python
def compare_models(filtered_feature_df, test_df):
    models = [smf.ols('weight ~ x + z', data=filtered_feature_df),
         smf.ols('weight ~ x*z', data=filtered_feature_df),
         smf.ols('log_weight ~ x+z', data=filtered_feature_df.loc[filtered_feature_df.weight>0])]

    for index, model in enumerate(models):
        eval_model = model.fit()
        if (index < 2): 
            print('Mean Relative Error:', evaluation_df(test_df, ['x', 'z'], eval_model))
            print()
        else:
            print('Mean Relative Error:', evaluation_df_log(test_df, ['x', 'z'], eval_model))
```

```python
full_data = generate_feature_df(filter_freq=None)
filtered_data = full_data.loc[(full_data.weight >= 50) & (full_data.weight <= 200)].reset_index(drop=True)
filtered_data['log_weight'] = np.log(filtered_data.weight)

compare_models(filtered_data, unfiltered_test_set)
```

We observe that the model with filtering actually performs worse in the test set than the other models that do not apply filtering. 

We hypothesize that the reason for this is the choice of filtering may be very sensitive to the accelerometer readings. Since the test set was recorded on a different day from the train set, there may have been slight (unnoticable) changes in the environment of testing, which thus may have led to fluctuations in the sensor output - leading to the filtering process being unreliable.
